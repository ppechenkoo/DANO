```python
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("https://pechenko.me/tinkoff_chashback.csv")
# df = pd.read_csv("tinkoff_chashback.csv")
```

```python
df = df.assign(
    Air_sum = [0 for i in range(len(df))],
    Beauty_sum = [0 for i in range(len(df))],
    Book_sum = [0 for i in range(len(df))],
    Children_sum = [0 for i in range(len(df))],
    Cinema_sum = [0 for i in range(len(df))],
    Clothes_sum = [0 for i in range(len(df))],
    Electronics_sum = [0 for i in range(len(df))],
    Fastfood_sum = [0 for i in range(len(df))],
    Fuel_sum = [0 for i in range(len(df))],
    House_sum = [0 for i in range(len(df))],
    Jewelry_sum = [0 for i in range(len(df))],
    Other_sum = [0 for i in range(len(df))],
    Pet_sum = [0 for i in range(len(df))],
    Pharmacies_sum = [0 for i in range(len(df))],
    Supermarkets_sum = [0 for i in range(len(df))],
    Transport_sum = [0 for i in range(len(df))],
    total_last_3_month = [0 for i in range(len(df))],
    total_before = [0 for i in range(len(df))],
    total = [0 for i in range(len(df))],
    cb_sum_last_3_month = [0 for i in range(len(df))],
    cb_sum_before = [0 for i in range(len(df))],
    cb_sum = [0 for i in range(len(df))],
    cnt_cat = [0 for i in range(len(df))],
    conversion = [0 for i in range(len(df))]
    )
df.total_last_3_month = df["Air"] + df["Beauty shop"] + df["Book store"] + df["Children"] + df["Cinema"] + df["Clothes"] + df["Electronics"] + df["Fastfood"] + df["Fuel"] + df["House"] + df["Jewelry"] + df["Other"] + df["Pet shop"] + df["Pharmacies"] + df["Supermarkets"] + df["Taxi"] + df["Transport"]
df.total_before = df["Air_before"] + df["Beauty shop_before"] + df["Book store_before"] + df["Children_before"] + df["Cinema_before"] + df["Clothes_before"] + df["Electronics_before"] + df["Fastfood_before"] + df["Fuel_before"] + df["House_before"] + df["Jewelry_before"] + df["Other_before"] + df["Pet shop_before"] + df["Pharmacies_before"] + df["Supermarkets_before"] + df["Taxi_before"] + df["Transport_before"]
df.total = df.total_last_3_month + df.total_before
df.cb_sum_last_3_month = df.cb_merch_last_3_month + df.cb_bank_last_3_month
df.cb_sum_before = df.cb_merch_before + df.cb_bank_before
df.cb_sum = df.cb_sum_last_3_month + df.cb_sum_before
df.Air_sum = df["Air"] + df["Air_before"]
df.Beauty_sum = df["Beauty shop"] + df["Beauty shop_before"]
df.Book_sum = df["Book store"] + df["Book store_before"]
df.Children_sum = df["Children"] + df["Children_before"]
df.Cinema_sum = df["Cinema"] + df["Cinema_before"]
df.Clothes_sum = df["Clothes"] + df["Clothes_before"]
df.Electronics_sum = df["Electronics"] + df["Electronics_before"]
df.Fastfood_sum = df["Fastfood"] + df["Fastfood_before"]
df.Fuel_sum = df["Fuel"] + df["Fuel_before"]
df.House_sum = df["House"] + df["House_before"]
df.Jewelry_sum = df["Jewelry"] + df["Jewelry_before"]
df.Other_sum = df["Other"] + df["Other_before"]
df.Pet_sum = df["Pet shop"] + df["Pet shop_before"]
df.Pharmacies_sum = df["Pharmacies"] + df["Pharmacies_before"]
df.Supermarkets_sum = df["Supermarkets"] + df["Supermarkets_before"]
df.Transport_sum = df["Transport"] + df["Transport_before"]
df.conversion = df.cnt_activation / df.cnt_view
df.drop(["Taxi", "Taxi_before"], axis= 1, inplace= True)
df.sort_values("total", inplace=True)
df.reset_index(inplace=True)
```

```python
df.plot(y = "total", kind="box", figsize=(3,6))
```

```python
df = df.query("total > 0")
df = df.query("total < 35992")
df = df.query("cnt_activation > 0" and "cnt_view > 0")
df.reset_index(inplace=True)
```

```python
categories_sum = ["Air_sum", "Beauty_sum", "Book_sum", "Children_sum", "Cinema_sum", "Clothes_sum", "Electronics_sum", "Fastfood_sum", "Fuel_sum", "House_sum", "Jewelry_sum", "Other_sum", "Pet_sum", "Pharmacies_sum", "Supermarkets_sum", "Transport_sum"]

for i in range(len(df)):
    temp = 0
    for c in categories_sum:
        if df[c][i] > 0:
            temp += 1
    df.cnt_cat[i] = temp
```

```python
first = df.query("cnt_communication == 0")
second = df.query("1 <= cnt_communication <= 2")
third = df.query("3 <= cnt_communication")
```

```python
comm_count = pd.DataFrame({
    "zero": first.mean(),
    "less": second.mean(),
    "many": third.mean()
})
comm_count = comm_count.T
comm_count = comm_count.assign(
    conversion_std = [first.conversion.std(), second.conversion.std(), third.conversion.std()],
    count = [len(first), len(second), len(third)],
    total_std = [first.total.std(), second.total.std(), third.total.std()],
    popularity = [first.cnt_cat.mean(), second.cnt_cat.mean(), third.cnt_cat.mean()],
    popularity_std = [first.cnt_cat.std(), second.cnt_cat.std(), third.cnt_cat.std()]
)
comm_count = comm_count.T
comm_count.drop(["client_id", "index", "level_0"], inplace = True)
comm_count = comm_count.T
comm_count
```

```python
comm_count.plot(y = ["total"],kind = "bar", grid=(1,1), figsize=(8,4), ylim=(10000))
```

```python
comm_count.plot(y = ["conversion"],kind = "bar", grid=(1,1), figsize=(4,5), ylim=(0.4))
```

```python
comm_count.plot(y = ["popularity"],kind = "bar", grid=(1,1), figsize=(8,4), ylim=(9))
```

```python
categories = ["Air_sum", "Beauty_sum", "Book_sum", "Children_sum", "Cinema_sum", "Clothes_sum", "Electronics_sum", "Fastfood_sum", "Fuel_sum", "House_sum", "Jewelry_sum", "Other_sum", "Pet_sum", "Pharmacies_sum", "Supermarkets_sum", "Transport_sum"]
categories_count = [0 for i in range(16)]

for i in range(len(df)):
    for n in range(len(categories)):
        c = categories[n]
        if df[c][i] > 0:
            categories_count[n] += 1

print(categories_count)
```
